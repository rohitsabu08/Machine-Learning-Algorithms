OLS REGRESSION MODEL 
From the cars data consider MPG as target and weight,HP,Displacement as predictors, Fit a LR model.


```python
#importing the libraries
import os
import numpy as np
import pandas as pd
```

```python
cars=pd.read_csv("cars.csv")  #importing the cars dataset
```

#### Y=MPG , X=Weight,HP, Displacement   

```python
y=cars.MPG                                              # Target Value
x=cars[["Weight" , "Horsepower" , 'Displacement']]      # Predictors
x1 = np.matrix(x)                                       # Used for further calc in y prediction

x.insert(0,"", pd.Series(np.ones(len(x))) , True)       # Inserting a column of ones into the dataframe
x_transpose =np.transpose(x)                            # To transpose the matrix


# Converting the pd dataframe to array for matrix calculation

x=np.matrix(x)
x_transpose=np.matrix(x_transpose)
y=np.array(y)


# Calc the b-curl values
 
temp = np.matmul(x_transpose,x)        #Step1 = X*X(transpose)
temp = np.linalg.inv(temp)             #Step2 = Inverse of temp
temp = np.matmul(temp,x_transpose)     #Step3 = Xly with x transpose again
temp = np.matmul(temp,y)               #Step4 = Xly with y
```

```python
#Function to calculate the y predicted value

def predict_y(x):
    y_pred=[]
    for i in range(0,406):
        val = temp[0,0] + x[i,0]*temp[0,1]+ x[i,1]*temp[0,2]+ x[i,2]*temp[0,3]
        y_pred.append(val)
    #print(y_pred)
    return(y_pred)
    
```

```python
y_predicted = predict_y(x1)          #Calling the function

print("Actual Y value \t Predicted Y value")
for i in range(0,406):
    print("   ",y[i],"\t",y_predicted[i])
```
